PCK

```python
y = []
y.append(mov_pck)
y.append(mov_pck)
```

```python
med_per = [0]*5
temp = set(med_pck)
for pck in temp:
    if pck == 0:
        med_per[0] = med_per[0] + med_pck.count(pck)
    if 0 < pck <= 0.25:
        med_per[1] = med_per[1] + med_pck.count(pck)
    if 0.25 < pck <= 0.5:
        med_per[2] = med_per[2] + med_pck.count(pck)
    if 0.5 < pck <= 0.75:
        med_per[3] = med_per[3] + med_pck.count(pck)
    if 0.75 < pck <= 1:
        med_per[4] = med_per[4] + med_pck.count(pck)  
med_per  
```

```python
mov_per = [0]*5
temp = set(mov_pck)
for pck in temp:
    if pck == 0:
        mov_per[0] = mov_per[0] + mov_pck.count(pck)
    if 0 < pck <= 0.25:
        mov_per[1] = mov_per[1] + mov_pck.count(pck)
    if 0.25 < pck <= 0.5:
        mov_per[2] = mov_per[2] + mov_pck.count(pck)
    if 0.5 < pck <= 0.75:
        mov_per[3] = mov_per[3] + mov_pck.count(pck)
    if 0.75 < pck <= 1:
        mov_per[4] = mov_per[4] + mov_pck.count(pck)  
mov_per  
```

```python
plt.figure().set_figwidth(10)
plt.title("PCK Comparison")
plt.xlim(0,3)
plt.ylim(-0.1,1)
plt.boxplot(y, widths = [0.5,0.5])
plt.ylabel("Percentage of Correct Keypoints (PCK)")
plt.grid(linestyle='-', linewidth=0.2)
ticks = range(1,3)
ticks_name = ['MediaPipe Pose', 'MoveNet']
plt.xticks(ticks, ticks_name)
plt.show()
```

Accuracy

```python
keypoints = range(1,18)
plt.figure().set_figwidth(12)
barwidth = 0.4
pos1 = [x - barwidth/2 for x in keypoints] 
pos2 = [x + barwidth/2 for x in keypoints] 
ticks = range(0,18)
# Plot the first set of bars
plt.bar(pos1, med_acc, color='blue', width = barwidth, label='MediaPipe Pose', alpha=0.7)

# Plot the second set of bars
plt.bar(pos2, mov_acc, color='red', width = barwidth, label='MoveNet', alpha=0.7)

# Add labels and title
plt.xlabel('Keypoints')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.legend()
plt.xticks(ticks)
plt.ylim(0,1)
plt.xlim(0, 18)
# Show the plot
plt.show()
```
